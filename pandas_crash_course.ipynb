## <font color=red>What Is a Pandas DataFrame?</font>

The pandas DataFrame object is a two-dimensional labled data structure with columns of the same or different data types; a column is a pandas Series object, which is a one-dimensional, labeled array made up of an autogenerated index and data. A DataFrame is like a sequence of Series objects aligned by the same index. **There are three main components that make up a pandas DataFrame: the index, the columns, and the data.**

The `.sample(n)` method returns a random sample of rows in the Series; n = 1 by default. Again, the index is retained.

## Indexing & Filtering

### The Indexing Operator `[]`

Above, I demonstrated that the bracket notation, `df['series']`, aka the indexing operator `[]`, can be used to pull a Series (column) from a pandas DataFrame when a string column name is passed in. Using the indexing operator can do more than just return a column or range of columns from a DataFrame. 

I can also filter a Series (column) or entire DataFrame using the indexing operator. 

I will demonstrate concrete examples of indexing and filtering using the indexing operator `[]` below.

````
# Let's grab a subset of the original dataset titanic_df. 
# Double brackets because I'm passing a list to the indexing operator.

titanic_df[['sex', 'age', 'pclass']]
````

````
# If I want to keep this subset DataFrame to use later, I have to assign it to a variable.

subset1 = titanic_df[['sex', 'age', 'pclass']]
subset1.head()
````

````
# We made a change; let's use our handy shape attribute! 891 rows and 3 columns.

subset1.shape
````

````
# I can filter my DataFrame based on a condition, too! Let's use subset1.

female_filter = subset1['sex'] == 'female'
female_filter
````

````
# Now the cool part! I can pass the mask into the indexing operator to the subset1 DataFrame.
# This returns only observations where my condition of sex == female is True.
# I can save this to a variable and have another subset of only female observations if I wanted.

females = subset1[female_filter]
females.sample(5)
````

````
# Ok, one more time. Our females DataFrame has only 314 rows and 3 columns.

females.shape
````

### `.value_counts()`

The `.value_counts()` method returns the frequency of occurance of the unique labels in the a column or Series. This is an extremely useful method you will find yourself using often with Series containing object and category data types.

````
# Survived is a binary column with 1 representing survivors; there are no null values.

titanic_df.survived.value_counts(dropna=False)
````

___

### `.astype()` & `.dtype`

The `.astype()` method allows me to convert a Series from one data type to another. Like most methods, it returns a new transformed Series by default instead of mutating my original data. `.dtype` attribute lets me check the data type.

````
# I could change a numeric column to be a string if it suited my needs. I have to reassign to keep it.

titanic_df.passenger_id = titanic_df.passenger_id.astype(object)
````

````
titanic_df.passenger_id.dtype
````
